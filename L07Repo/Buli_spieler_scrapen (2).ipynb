## Erster Spieler vor und Nachname

```python
import requests
from lxml import html

# URL der Website
url = 'https://www.bundesliga.com/de/bundesliga/spieler'

# Sende eine GET-Anfrage an die Website
response = requests.get(url)

# Überprüfe den Statuscode der Antwort
if response.status_code == 200:
    # Parse den HTML-Inhalt der Seite
    tree = html.fromstring(response.content)
    
    # Definiere den XPath
    xpath = '/html/body/bundesliga-root/wrapper-page/div/div/div/div/div/player-overview-page/div/article/mat-accordion/mat-expansion-panel[1]/div/div/div[1]/div[2]/player-card-simple/div/a/div/span[2]'
    
    # Finde alle Elemente, die dem XPath entsprechen
    players = tree.xpath(xpath)
    
    # Extrahiere und drucke die Spielernamen
    for player in players:
        print(player.text_content().strip())
else:
    print(f"Fehler beim Abrufen der Seite: {response.status_code}")
```

## Vor und Nachnamen

```python
import requests
from lxml import html

# URL der Website
url = 'https://www.bundesliga.com/de/bundesliga/spieler'

# Sende eine GET-Anfrage an die Website
response = requests.get(url)

# Überprüfe den Statuscode der Antwort
if response.status_code == 200:
    # Parse den HTML-Inhalt der Seite
    tree = html.fromstring(response.content)
    
    # XPath für die Container aller Spieler
    xpath_player_cards = '//player-card-simple'
    
    # Finde alle Spieler-Karten
    player_cards = tree.xpath(xpath_player_cards)
    
    for card in player_cards:
        # XPath für Vorname und Nachname innerhalb der Spieler-Karte
        vorname_xpath = './/span[2]/span[1]/text()'
        nachname_xpath = './/span[2]/span[2]/text()'
        
        # Extrahiere Vor- und Nachnamen
        vorname = card.xpath(vorname_xpath)
        nachname = card.xpath(nachname_xpath)
        
        # Überprüfen, ob Vorname und Nachname vorhanden sind
        if vorname and nachname:
            print(f"{vorname[0].strip()} {nachname[0].strip()}")
        else:
            print("Fehler: Vorname oder Nachname fehlen.")
else:
    print(f"Fehler beim Abrufen der Seite: {response.status_code}")
```

## Anzahl der spieler und deren vor und nachnamen

```python
import requests
from lxml import html

# URL der Website
url = 'https://www.bundesliga.com/de/bundesliga/spieler'

# Sende eine GET-Anfrage an die Website
response = requests.get(url)

# Überprüfe den Statuscode der Antwort
if response.status_code == 200:
    # Parse den HTML-Inhalt der Seite
    tree = html.fromstring(response.content)
    
    # XPath für die Container aller Spieler
    xpath_player_cards = '//player-card-simple'
    
    # Finde alle Spieler-Karten
    player_cards = tree.xpath(xpath_player_cards)
    
    # Anzahl der Spieler-Karten
    anzahl_spieler = len(player_cards)
    
    print(f"Anzahl der Spieler auf der Website: {anzahl_spieler}")
    
    for card in player_cards:
        # XPath für Vorname und Nachname innerhalb der Spieler-Karte
        vorname_xpath = './/span[2]/span[1]/text()'
        nachname_xpath = './/span[2]/span[2]/text()'
        
        # Extrahiere Vor- und Nachnamen
        vorname = card.xpath(vorname_xpath)
        nachname = card.xpath(nachname_xpath)
        
        # Überprüfen, ob Vorname und Nachname vorhanden sind
        if vorname and nachname:
            print(f"{vorname[0].strip()} {nachname[0].strip()}")
        else:
            print("Fehler: Vorname oder Nachname fehlen.")
else:
    print(f"Fehler beim Abrufen der Seite: {response.status_code}")
```
