#### Frage 1: Welche Nationalitäten (aus Tabelle 1) haben Spieler mit demselben Vornamen wie Personen in Tabelle 2? Hier wird eine Verbindung zwischen Vornamen aus Tabelle 2 und Nationalitäten aus Tabelle 1 hergestellt.

```python
df_csv['vorname_spieler'] = df_csv['name'].str.split().str[0]
df_csv_vornamen = df_csv['vorname_spieler']
print(df_csv_vornamen)
```

```python
# Verknüpfen der beiden Tabellen basierend auf dem Vornamen
result = pd.merge(df_csv, df_json, left_on='vorname_spieler', right_on='vorname')
print(result)
```

```python
# Entferne Duplikate, um einzigartige Kombinationen aus Namen und Nationalität zu bekommen
unique_nationalities = result[["name", "nationality"]].drop_duplicates()

# Ausgabe der Namen und Nationalitäten
print(unique_nationalities)
```

### Frage 2: Wie viele Spieler aus Deutschland haben denselben Vornamen wie eine weibliche Person in der zweiten Tabelle? Hier könnte man Nationalität aus der ersten Tabelle und das Geschlecht und den Vornamen aus der zweiten Tabelle verknüpfen.

```python
# 1. Vorname der Spieler aus df_csv extrahieren
df_csv['vorname_spieler'] = df_csv['name'].str.split().str[0]
print(df_csv['vorname_spieler'])
```

```python
# 2. Filtere Spieler, die aus Deutschland kommen
deutsche_spieler = df_csv[df_csv['nationality'] == 'Germany']
print(deutsche_spieler)
```

```python
# 3. Filtere weibliche Personen aus df_json
weibliche_personen = df_json[df_json['geschlecht'] == 'w']['vorname']
print(weibliche_personen)
```

```python
# 4. Verknüpfe Spieler mit denselben Vornamen wie weibliche Personen
matching_spieler = deutsche_spieler[deutsche_spieler['vorname_spieler'].isin(weibliche_personen)]
print(matching_spieler)
```

```python
# 5. Anzahl der passenden Spieler zählen
anzahl_matching_spieler = matching_spieler.shape[0]
print(anzahl_matching_spieler)
```

```python
# Ausgabe der Anzahl der passenden Spieler
print(f"Anzahl der Spieler aus Deutschland mit einem Vornamen, der auch bei weiblichen Personen vorkommt: {anzahl_matching_spieler}")
```

```python
# 1. Namen der Spieler, die mit weiblichen Vornamen übereinstimmen, anzeigen
spielernamen = matching_spieler['name']

# 2. Ausgabe der Namen
print("Namen der Spieler, deren Vorname mit weiblichen Personen übereinstimmt:")
print(spielernamen)
```
